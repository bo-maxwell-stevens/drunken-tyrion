```python
def getPDFContent(path):
    content = ""
    # Load PDF into pyPDF
    pdf = pyPdf.PdfFileReader(file(path, "rb"))
    # Iterate pages
    for i in range(0, pdf.getNumPages()):
        # Extract text from page and add to content
        content += pdf.getPage(i).extractText() + "/n"
    # Collapse whitespace
    content = " ".join(content.replace(u"/xa0", " ").strip().split())
    return content
```

```python
filelist = {}

filename_list = ['1.pdfx', '2.pdfx', '3.pdfx']

for files in filename_list:
    filename = os.path.abspath(files)
    
    filelist[files] = getPDFContent(filename).encode("ascii", "ignore")
    filelist[files] = filelist[files].split('References')[1]

    m = re.findall(r'\([0-9]{4}\)(.+?)(?:http\:|\([0-9]+|(?:\:[0-9]+?))', filelist[files])

    new_list = []

    for i in range(len(m)):
        found = re.search(r'^([^.]+).*', m[i])
        if found:
            if len(found.group(1)) > 10:
                string = found.group(1)
                string = ''.join(e for e in string if e.isalnum())
                new_list.append(string)
    filelist[files] = new_list
```

```python
def compare_dist(first, second):
    return distance.levenshtein(first, second, normalized=True)
```
